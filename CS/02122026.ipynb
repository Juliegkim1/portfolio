1. Reverse a Doubly LInked List
Write a function to reverse the elements in a doubly linked list. Do not simply print it out, replace the values or use methods. It must have the references correctly set in reversed order. 
Also, write a print method – this should help with debugging.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None


class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
    
    def append(self, data):
        """Add a node to the end of the list"""
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            self.tail = new_node
        else:
            new_node.prev = self.tail
            self.tail.next = new_node
            self.tail = new_node
    
    def reverse(self):
        """Reverse the doubly linked list by swapping next and prev pointers"""
        if not self.head:
            return  # Empty list
        
        current = self.head
        
        # Swap next and prev pointers for each node
        while current:
            # Swap the next and prev pointers
            temp = current.next
            current.next = current.prev
            current.prev = temp
            
            # Move to the next node 
            current = temp
        
        # Swap head and tail
        temp = self.head
        self.head = self.tail
        self.tail = temp
    
    def print_forward(self):
        """Print the list from head to tail"""
        if not self.head:
            print("List is empty")
            return
        
        current = self.head
        elements = []
        while current:
            elements.append(str(current.data))
            current = current.next
        
        print(" <-> ".join(elements))
    
    def print_backward(self):
        """Print the list from tail to head"""
        if not self.tail:
            print("List is empty")
            return
        
        current = self.tail
        elements = []
        while current:
            elements.append(str(current.data))
            current = current.prev
        
        print(" <-> ".join(elements))
```

2. Rotate a Singly Linked List

Write a function that rotates a singly linked list.

If n is positive, rotate the list by n positions to the right
If n is negative, rotate the list by abs(n) positions to the left
Make sure the function is efficient and do not convert the linked list into another data structure
Finally, clearly state the time complexity and additional space complexity of your solution
 

Examples:
1->2->3->4->5 (rotate 2) result: 4->5->1->2->3
1->2->3->4->5 (rotate -2) result: 3->4->5->1->2

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class SinglyLinkedList:
    def __init__(self):
        self.head = None
    
    def append(self, data):
        """Add a node to the end of the list"""
        new_node = Node(data)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
    
    def rotate(self, n):
        """
        Rotate the linked list by n positions.
        """
        if not self.head or not self.head.next:
            return  # Empty list or single node
        
        # Step 1: Find the length of the list and get the tail
        length = 1
        tail = self.head
        while tail.next:
            tail = tail.next
            length += 1
        
        # Step 2: Normalize n to be within [0, length)
        n = n % length
        
        if n == 0:
            return #no rotation needed for 0
        
        # Step 3: Convert right rotation to left rotation
        if n > 0:
            n = length - n
        else:
            n = abs(n)
        
        # Step 4: Find the new tail- node at position n-1
        new_tail = self.head
        for _ in range(n - 1):
            new_tail = new_tail.next
        
        # Step 5: Perform the rotation
        tail.next = self.head
        
        # Update to the new head
        self.head = new_tail.next
        
        # Break the circle at the new tail
        new_tail.next = None
    
    def print_list(self):
        """Print the list"""
        if not self.head:
            print("List is empty")
            return
        
        current = self.head
        elements = []
        while current:
            elements.append(str(current.data))
            current = current.next
        
        print(" -> ".join(elements))
```

```python
print("Test 1: Rotate right by 2")
ll1 = SinglyLinkedList()
for i in range(1, 6):
    ll1.append(i)

print("Original list:")
ll1.print_list()

ll1.rotate(2)
print("After rotating right by 2:")
ll1.print_list()
print()

print("Test 2: Rotate left by 2")
ll2 = SinglyLinkedList()
for i in range(1, 6):
    ll2.append(i)

print("Original list:")
ll2.print_list()

ll2.rotate(-2)
print("After rotating left by 2:")
ll2.print_list()
print()

print("Test 3: Rotate by length")
ll3 = SinglyLinkedList()
for i in range(1, 6):
    ll3.append(i)

print("Original list:")
ll3.print_list()

ll3.rotate(5)
print("After rotating right by 5:")
ll3.print_list()
print()

print("Test 4: Rotate by more than length")
ll4 = SinglyLinkedList()
for i in range(1, 6):
    ll4.append(i)

print("Original list:")
ll4.print_list()

ll4.rotate(7)  
print("After rotating right by 7 (equivalent to 2):")
ll4.print_list()
print()

print("Test 5: single node")
ll5 = SinglyLinkedList()
ll5.append(1)

print("Original list:")
ll5.print_list()

ll5.rotate(3)
print("After rotating by 3:")
ll5.print_list()
```

TIME COMPLEXITY IS O(N)  AND THE SPACE COMPLEXITY IS O(1) AS IT USES A FIXED NUMBER OF VARIABLES.

3. Linked List Union

Write an efficient function that takes two linked lists as input and returns a new linked list containing their union (all distinct elements from both lists).

Ensure the implementation is as space and time efficient as possible
The lists are not necessarily sorted
The order of the output is not significant
Do not use built-in union operators or functions
For example:

[2, 10, 5, 3, 4] and [4, 7, 8, 3, 11] has a union of [2, 10, 3, 4, 5, 7, 8, 11]

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class SinglyLinkedList:
    def __init__(self):
        self.head = None
    
    def append(self, data):
        """Add a node to the end of the list"""
        new_node = Node(data)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
    
    def print_list(self):
        if not self.head:
            print("List is empty")
            return
        
        current = self.head
        elements = []
        while current:
            elements.append(str(current.data))
            current = current.next
        
        print(" -> ".join(elements))

# A new SinglyLinkedList containing all distinct elements
def union(list1, list2):
    result = SinglyLinkedList()
    
    # Hash set to track
    seen = set()
    
    # Add all elements from list1
    current = list1.head
    while current:
        if current.data not in seen:
            seen.add(current.data)
            result.append(current.data)
        current = current.next
    
    # Add all elements from list2 that aren't already in the set
    current = list2.head
    while current:
        if current.data not in seen:
            seen.add(current.data)
            result.append(current.data)
        current = current.next
    
    return result
```

```python
ll1 = SinglyLinkedList()
for val in [2, 10, 5, 3, 4]:
    ll1.append(val)

ll2 = SinglyLinkedList()
for val in [4, 7, 8, 3, 11]:
    ll2.append(val)

print("List 1:")
ll1.print_list()
print("\nList 2:")
ll2.print_list()

result = union(ll1, ll2)
print("\nUnion:")
result.print_list()
print()
```

4. Flatten a Binary Tree

Write two functions to convert a valid binary tree, flatten it into a linked list and return it.

The first function should flatten it using preorder traversal
The second function should flatten it using postorder traversal
Write it as time and space efficient as possible.
Preorder linked list: 1->2->4->5->3->6
Postorder linked list: 4->5->2->6->3->1

```python
class TreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


class LinkedListNode:
    def __init__(self, data):
        self.data = data
        self.next = None


class LinkedList:
    def __init__(self):
        self.head = None
    
    def print_list(self):
        if not self.head:
            print("List is empty")
            return
        
        current = self.head
        elements = []
        while current:
            elements.append(str(current.data))
            current = current.next
        
        print(" -> ".join(elements))


def flatten_preorder(root):
    result = LinkedList()
    
    if not root:
        return result    

    tail = [None] 
    
    def preorder_helper(node):
        if not node:
            return

        new_node = LinkedListNode(node.data)
        
        # Add to linked list
        if not result.head:
            result.head = new_node
            tail[0] = new_node
        else:
            tail[0].next = new_node
            tail[0] = new_node
        
        # Recursively process 
        preorder_helper(node.left)
        preorder_helper(node.right)
    
    preorder_helper(root)
    return result


def flatten_postorder(root):   
    result = LinkedList()
    
    if not root:
        return result

    tail = [None]  
    
    def postorder_helper(node):
        if not node:
            return
        
        # Recursively process 
        postorder_helper(node.left)
        postorder_helper(node.right)
        
        new_node = LinkedListNode(node.data)
        
        # Add to linked list
        if not result.head:
            result.head = new_node
            tail[0] = new_node
        else:
            tail[0].next = new_node
            tail[0] = new_node
    
    postorder_helper(root)
    return result


def flatten_preorder_iterative(root):
    result = LinkedList()
    
    if not root:
        return result
    
    stack = [root]
    tail = None
    
    while stack:
        # Pop node from stack
        node = stack.pop()
        
        # Create linked list node
        new_node = LinkedListNode(node.data)
        
        # Add to linked list
        if not result.head:
            result.head = new_node
            tail = new_node
        else:
            tail.next = new_node
            tail = new_node
        
        # Push right child first
        if node.right:
            stack.append(node.right)
        if node.left:
            stack.append(node.left)
    
    return result


def flatten_postorder_iterative(root):    
    result = LinkedList()
    
    if not root:
        return result
    
    stack1 = [root]
    stack2 = []
    
    # Root -> Right -> Left into stack2
    while stack1:
        node = stack1.pop()
        stack2.append(node)
        
        # right is processed first
        if node.left:
            stack1.append(node.left)
        if node.right:
            stack1.append(node.right)
    
    # linked list from stack2
    tail = None
    while stack2:
        node = stack2.pop()

        new_node = LinkedListNode(node.data)
        
        # Add to linked list
        if not result.head:
            result.head = new_node
            tail = new_node
        else:
            tail.next = new_node
            tail = new_node
    
    return result
```

```python
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.right = TreeNode(6)
```

```python
# Test preorder flattening 
print("\nPreorder Flattening")
print("Expected: 1 -> 2 -> 4 -> 5 -> 3 -> 6")
print("Result:   ", end="")
preorder_list = flatten_preorder(root)
preorder_list.print_list()
```

```python
print("\nPostorder Flattening")
print("Expected: 4 -> 5 -> 2 -> 6 -> 3 -> 1")
print("Result:   ", end="")
postorder_list = flatten_postorder(root)
postorder_list.print_list()
```

```python
print("\nPreorder Flattening - Iterative")
print("Expected: 1 -> 2 -> 4 -> 5 -> 3 -> 6")
print("Result:   ", end="")
preorder_list_iter = flatten_preorder_iterative(root)
preorder_list_iter.print_list()
```

```python
print("\nPostorder Flattening - Iterative")
print("Expected: 4 -> 5 -> 2 -> 6 -> 3 -> 1")
print("Result:   ", end="")
postorder_list_iter = flatten_postorder_iterative(root)
postorder_list_iter.print_list()
```

For iterative, time complexity is the same as O(n) but the space compexity is different for pre and post; O(h) and O(n), respectively.
