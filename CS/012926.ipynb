1. Write a function that accepts a deque and an integer, n

It should then rotate the contents of the deque by n steps to the right
If n is negative, it should rotate it to the left
Do not use the built-in rotate functions/methods
You may write your own deque class, use the one from the dsa module (dsa.deque) or use the standard Python dequeLinks to an external site.
Provide the Big O time complexity for both the average and worst-case scenarios
Provide the Big O space complexity, considering only the additional space required beyond the input

```python
from collections import deque

def rotate_deque(dq, n):
    if not dq or n == 0:
        return

    length = len(dq)
    n = n % length 
    
    # If n becomes 0 after, do nothing
    if n == 0:
        return

    # pop right, append left
    for _ in range(n):
        item = dq.pop()      
        dq.appendleft(item)  
```

```python
d1 = deque([1, 2, 3, 4, 5])
rotate_deque(d1, 2)
print(f"Right Rotation (2): {d1}")
d2 = deque([1, 2, 3, 4, 5])
rotate_deque(d2, -1) 
print(f"Left Rotation (-1): {d2}")  
```

2. Reverse a Queue

Write a function that accepts a queue and reverses its contents. It should only use a queue data structure and recursion to reverse the queue in place.

You may use Python’s collections.deque as long as you treat it strictly as a queue.

```python
from collections import deque

def reverse_queue(q):
    
    if not q:
        return

    # Step 1: Dequeue the front element
    front_element = q.popleft()

    # Step 2: Recursive - remaining elements
    reverse_queue(q)

    # Step 3: append the element to the back
    q.append(front_element)

```

```python
my_queue = deque([1, 2, 3, 4, 5])
print(f"Original Queue: {my_queue}")

reverse_queue(my_queue)    
print(f"Reversed Queue: {my_queue}")
```

3. Recursive Binary Search

Write a recursive binary search function to search for an element in an array. Assume the elements in the array are sorted.

It should return the index of the element and return -1 if it is not found.

```python
def binary_search_recursive(arr, target, low, high):
    # Base Case
    if low > high:
        return -1

    # Calculate the middle 
    mid = (low + high) // 2

    if arr[mid] == target:
        return mid

    elif target < arr[mid]:
        return binary_search_recursive(arr, target, low, mid - 1)

    else:
        return binary_search_recursive(arr, target, mid + 1, high)

# Helper function 
def binary_search(arr, target):
    if not arr:
        return -1
    return binary_search_recursive(arr, target, 0, len(arr) - 1)
```

```python
sorted_list = [2, 5, 8, 12, 16, 23, 38, 56, 72, 91]
search_target = 23
    
index = binary_search(sorted_list, search_target)

print(f"Looking for {search_target} in {sorted_list}")
print(f"Found at index: {index}") 
```

4. Balanced Brackets

Write a function that returns a Boolean indicating whether a string has balanced brackets. You may ignore non-bracket characters. For our purposes, the following are bracket characters:

(){}[]
Balanced brackets refer to a string where every opening bracket has a corresponding closing bracket, and the pairs are properly nested.

For example, this function should return True for the following: 

[ ]
{}{}[]()
[{()}]
(()[[[()({})]]])
It should return False for the following:

[ ] [
{{}[](})
[{)}]
(()[()({})]]])

```python
def is_balanced(s):
    matching_bracket = {')': '(', '}': '{', ']': '['}
    
    # We use a list as a stack (append() for push, pop() for pop)
    stack = []
    
    for char in s:
        if char in matching_bracket.values(): # '(', '{', '['
            stack.append(char)
            
        elif char in matching_bracket:
            if not stack:
                return False          

            top_element = stack.pop()
            if top_element != matching_bracket[char]:
                return False

    return len(stack) == 0

```

```python
print(is_balanced("[ ] [")) 
print(is_balanced("{{}[](})"))            
print(is_balanced("[{)}]"))               
print(is_balanced("(()[()({})]]])"))      
```

5. Postfix Notation

Write a function to evaluate an arithmetic expression in string format. The expression is in postfix notation. This means that:

Also known as Reverse Polish Notation (RPN), operators follow their operands
Example: The infix expression 3 + 4 is written as 3 4 + in postfix notation
Each operator follows its two operands
No Parentheses Needed: Postfix notation eliminates the need for parentheses to indicate operation precedence
The order of operations is strictly defined by the position of operators and operands
Assumptions:

The strings are syntactically correct
Operands and operators are separated by spaces
The function should handle the operators +, -, *, and /
Numbers can be integers or floating-points
Examples:

Input: "3 4 +" Output: 7.0
Input: "10 2 /" Output: 5.0
Input: "3 2 + 2 /" Output: 2.5
Input: "5 1 2 + 4 * + 3 -" Output: 14.0
Evaluation steps:
5 1 2 + 4 * + 3 -
5 3 4 * + 3 -
5 12 + 3 -
17 3 -
14.0


```python
def evaluate_postfix(exp):
    stack = []
    tokens = exp.split()
    
    for token in tokens:
        if token in "+-*/":
            right = stack.pop()
            left = stack.pop()
            
            if token == '+':
                result = left + right
            elif token == '-':
                result = left - right
            elif token == '*':
                result = left * right
            elif token == '/':

                result = left / right
                
            stack.append(result)            
             
    return stack[0]
```

```python
print(f"3 4 + -> {evaluate_postfix('3 4 +')}")
print(f"10 2 / -> {evaluate_postfix('10 2 /')}")
print(f"3 2 + 2 / -> {evaluate_postfix('3 2 + 2 /')}")

complex_ex = "5 1 2 + 4 * + 3 -"
print(f"{complex_ex} -> {evaluate_postfix(complex_ex)}") 
```

```python
def evaluate_postfix(exp):
    stack = []
    tokens = exp.split()
    
    for token in tokens:
        if token in "+-*/":
            right = stack.pop()
            left = stack.pop()
            
            if token == '+':
                result = left + right
            elif token == '-':
                result = left - right
            elif token == '*':
                result = left * right
            elif token == '/':

                result = left / right                
            stack.append(result)       
        else:
            stack.append(float(token))     
             
    return stack[0]
```
