#3. Rotate an Array
Write a function that performs a circular rotation on an array. The function should meet the following requirements:
Function Inputs:
An array of elements
An integer representing the number of positions to rotate:
A positive number rotates the array to the right
A negative number rotates the array to the left
Behavior:
The rotation should be circular, meaning elements that move past one end of the array reappear at the other end
The function should not use any built-in rotation methods (e.g., rotate(), deque.rotate(), etc.)
Performance:
Ensure the function is efficient and handles valid inputs of various sizes
Provide the Big O time complexity for both the average and worst-case scenarios
Provide the Big O space complexity, considering only the additional space required beyond the input
Examples:

array = [1, 2, 3, 4, 5]
rotation = 2
output: [4, 5, 1, 2, 3]

array = [1, 2, 3, 4, 5]
rotation = -2
output: [3, 4, 5, 1, 2]

```python
def rotate(array, k):
    n = len(array)
    k = k % n

    return array[-k:] + array[:-k]
rotated_array = rotate(array, 3)
print(rotated_array)
```

```python
rotated_array = rotate(array, -1)
print(rotated_array)
```

#4. Count Value in an Array

Write a function that accepts a sorted array of integers and a target value and returns the count of the target value.

The array may contain duplicates.
Ensure the function handles a variety of valid inputs.
Full credit for a O(log n) solution and partial credit otherwise.
Examples:

array = [0, 1, 2, 3, 3, 3, 3, 3, 3, 4, 4, 5] 
target = 3
output: 6

array = [1, 1, 2, 2, 4, 4, 5, 5, 5] 
target = 5
output: 3

array = [1, 1, 2, 3, 3, 4, 5, 5, 5]
target = 6
output: 0

```python
def count_occurrences(arr, target):

    l = 0
    r = len(arr) - 1
    start_index = -1
    
    while l <= r:
        mid = (l + r) // 2
        if arr[mid] == target:
            start_index = mid
            r = mid - 1  
        elif arr[mid] < target:
            l = mid + 1
        else:
            r = mid - 1
            
    # If we didn't find it, we can just return 0 now
    if start_index == -1:
        return 0

    # the last instance
    l = 0
    r = len(arr) - 1
    end_index = -1
    
    while l <= r:
        mid = (l + r) // 2
        if arr[mid] == target:
            end_index = mid
            l = mid + 1  
        elif arr[mid] < target:
            l = mid + 1
        else:
            r = mid - 1
            
    return end_index - start_index + 1
```

```python
array = [0, 1, 2, 3, 3, 3, 3, 3, 3, 4, 4, 4, 5, 5] 
print(count_occurrences(array, 6))
```

#5. Maximum Value in a Rotated Array

Write a function that accepts a sorted, rotated array and returns the index of the largest value.

Assume that the array is sorted in ascending order (You may use the rotate() function that you wrote from #3 and any built-in sort() functions if necessary)
Full credit for a O(log n) solution and partial credit otherwise
Ensure that your function can handle all valid input types
Examples:

array = [4, 5, 1, 2, 3] 
max value index: 1 (value: 5)

array = [5, 7, 11, 0, 1, 3]
max value index: 2 (value: 11)

```python
def find_max_index(arr):
    if not arr:
        return -1
        
    n = len(arr)
    l = 0
    r = n - 1
    
    # Check if the array is already sorted (not rotated)
    if arr[l] <= arr[r]:
        return r
    
    while l <= r:
        mid = (l + r) // 2
        
        # Check if mid is the peak (the max value)
        if mid > 0 and arr[mid] < arr[mid - 1]:
            return mid - 1
        elif mid < n - 1 and arr[mid] > arr[mid + 1]:
            return mid
        elif arr[l] <= arr[mid]:
            # left side  sorted, 
            l = mid + 1
        else:
            # NOT sorted 
            r = mid - 1
            
    return -1 
```

```python
arr1 = [4, 5, 1, 2, 3]
idx1 = find_max_index(arr1)
print(f"Array: {arr1}")
print(f"Max Index: {idx1} (Value: {arr1[idx1] if idx1 != -1 else 'N/A'})\n")
arr2 = [6, 7, 8, 9, 10, 1, 2, 3, 4, 5]
idx2 = find_max_index(arr2)
print(f"Array: {arr2}")
print(f"Max Index: {idx2} (Value: {arr2[idx2] if idx2 != -1 else 'N/A'})\n")
```
